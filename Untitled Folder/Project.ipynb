```python
from tkinter import ttk
from tkinter import *
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd




root = Tk()
root.title("Machine Learning Classifier")
data_set = pd.read_excel('heart_disease.xlsx')
df= pd.DataFrame(data_set)
target = df['target']
features =df.drop(['target'],axis = 1) 
height = 30
width = 2

windowWidth = root.winfo_reqwidth()
windowHeight = root.winfo_reqheight()
positionRight = int(root.winfo_screenwidth()/3 - windowWidth/3)
positionDown = int(root.winfo_screenheight()/3 - windowHeight/2)
root.geometry('580x580')
root.geometry("+{}+{}".format(positionRight, positionDown))

top_frame =Frame(root, height = 600)
top_frame.pack(fill = X)
bottum_frame =Frame(root)
bottum_frame.pack(side = BOTTOM)

label_one = Label(top_frame, text = "Heart Disease Classifier" ,font=("Helvetica",  30), foreground = "#563A71")
label_one.pack()

canvas = Canvas(root,width = 50,height = 300,bg = "#8CCCC0")
canvas.pack(fill = X,ipady = 10)

labelTop = Label(canvas, text = "Choose your Algorithm",font=("Helvetica",  13), foreground = "#1762a3")
labelTop.pack(pady = 10 )

comboExample = ttk.Combobox(canvas, 
                            values=[
                                    "Neural Network", 
                                    "Support Vector Machines",
                                    "Decision Tree",
                                  "Naïve Bayes"], 
                                   width = 25,height = 30       )
comboExample.pack(pady = 6,padx = 6)
comboExample.current(0)

accurancy_label = Label(canvas, text = "  Accurancy  ",font=("Helvetica",  20), foreground = "#1762a3")
accurancy_label.pack(pady = 6)
pred_label = Label(canvas,font=("Helvetica",  22), foreground = "black")
pred_label.pack(pady = 6)

def PredictionWithNeuralNetwork(x_train, y_train, x_test, y_test ):
        neuralNetworkClassifier = MLPClassifier(hidden_layer_sizes = (2,5)) # neuron number 
        neuralNetworkClassifier.fit(x_train, y_train)
        y_pred = neuralNetworkClassifier.predict(x_test)
        return accuracy_score(y_test, y_pred)

def PredictionWithSVC(x_train, y_train, x_test, y_test ):
    svcClassifier = SVC(kernel='linear')
    svcClassifier.fit(x_train,y_train)  
    y_pred = svcClassifier.predict(x_test)
    return accuracy_score(y_test,  y_pred)

def PredictionWithGaussianNB(x_train, y_train, x_test, y_test ):
    gnb = GaussianNB()
    gnb.fit(x_train, y_train)
    y_pred = gnb.predict(x_test) 
    return accuracy_score(y_test, y_pred) 

def PredictionWithDecisionTree(x_train, y_train, x_test, y_test ):
    decisionTreeclf = DecisionTreeClassifier(criterion="entropy")
    decisionTreeclf.fit(x_train, y_train)
    y_pred = decisionTreeclf.predict(x_test) 
    return accuracy_score(y_test, y_pred) 
          
def GetAccurancy():
    x_train ,x_test ,y_train , y_test = train_test_split(features,target)
    if comboExample.get() == "Neural Network":
        pred_label.configure(text = str(PredictionWithNeuralNetwork(x_train, y_train, x_test, y_test )))
        
    elif  comboExample.get() == "Support Vector Machines":
         pred_label.configure(text = str(PredictionWithSVC(x_train, y_train, x_test, y_test )))
            
    elif  comboExample.get() == "Decision Tree":
        pred_label.configure(text = str(PredictionWithDecisionTree(x_train, y_train, x_test, y_test )))
        
    elif  comboExample.get() == "Naïve Bayes":
        pred_label.configure(text = str(PredictionWithGaussianNB(x_train, y_train, x_test, y_test )) )   
        
button_two = Button(canvas, text = "Predict" ,fg = "black" ,bg = "#3A7171"  ,padx =32,font=("Helvetica",  20)
                   , command = GetAccurancy)
button_two.pack()

def ShowData():
    window = Toplevel(root)
    age_label = Label(window,text = "age" )
    age_label.grid(row = 0, column = 0)
    
    sex_label = Label(window,text = "sex" )
    sex_label.grid(row = 0, column = 1)
    
    cp_label = Label(window,text = "cp" )
    cp_label.grid(row = 0, column = 2)
    
    trestbps_label = Label(window,text = "trestbps" )
    trestbps_label.grid(row = 0, column = 3)
    
    chol_label = Label(window,text = "chol" )
    chol_label.grid(row = 0, column = 4)
    
    fbs_label = Label(window,text = "fbs" )
    fbs_label.grid(row = 0, column = 5)
    
    restecg_label = Label(window,text = "restecg" )
    restecg_label.grid(row = 0, column = 6)

    for i in range(height): #Rows
                b = Label(window, text= str(features.age[i]))
                b.grid(row=i+1, column= 0)
               
                b1 = Label(window, text= str(features.sex[i]))
                b1.grid(row=i+1, column= 1)
                
                b4 = Label(window, text= str(features. cp[i]))
                b4.grid(row=i+1, column= 2)
                
                b2 = Label(window, text= str(features.trestbps[i]))
                b2.grid(row=i+1, column= 3)
                
                b3 = Label(window, text= str(features.chol[i]))
                b3.grid(row=i+1, column= 4)
                
                b5 = Label(window, text= str(features.fbs[i]))
                b5.grid(row=i+1, column= 5)
                
                b6 = Label(window, text= str(features.restecg[i]))
                b6.grid(row=i+1, column= 6)
                
               

        
button_one = Button(canvas, text = "Show Data" ,fg = "black" ,bg = "#3A7171"  ,padx = 12,font=("Helvetica",  20), command = ShowData)
button_one.pack(pady = 3)





root.mainloop()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd

data = pd.read_excel('heart_disease.xlsx')
df= pd.DataFrame(data)
Y = df['target']
X =df.drop(['target'],axis = 1) 
x_train ,x_test ,y_train , y_test = train_test_split(X,Y)


neuralNetworkClassifier = MLPClassifier(hidden_layer_sizes = (2,5)) # neuron number 
neuralNetworkClassifier.fit(x_train, y_train)
y_predict_neuralNetwork= neuralNetworkClassifier.predict(x_test)

svcClassifier = SVC(kernel='linear')
svcClassifier.fit(x_train, y_train)
y_predict_SVC = svcClassifier.predict(x_test)

gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_predict_NaiveBayes = gnb.predict(x_test)

decisionTreeclf = DecisionTreeClassifier()
decisionTreeclf.fit(x_train, y_train)
y_predict_DecisionTree = decisionTreeclf.predict(x_test)

print('NeuralNetwork accurancy = ',accuracy_score(y_test, y_predict_neuralNetwork))
print('SVC accurancy = ',accuracy_score(y_test, y_predict_SVC))
print('Naive Bayes accurancy =',accuracy_score(y_test, y_predict_NaiveBayes))
print('DecisionTree accurancy = ',accuracy_score(y_test, y_predict_DecisionTree))
```
